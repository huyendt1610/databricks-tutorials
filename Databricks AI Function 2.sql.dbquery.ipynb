```sql
-- 0. Create catalog & schema 
create catalog if not exists datasciencebasics; 
create schema if not exists datasciencebasics.basic; 

-- 1. Create the table 
create table if not exists datasciencebasics.basic.product_reviews(
  id int primary key, 
  review string
); 

-- 2. Insert data 
INSERT INTO datasciencebasics.basic.product_reviews(id, review) VALUES 
(1, 'This is a great product!'), 
(2, 'This is a bad product and I will never buy it again'), 
(3, 'This is a good product but it is too expensive'), 
(4, 'This gadget stopped working entirely after only two weeks. Total waste of money.'), 
(5, 'Tried contacting support three times with no response. Awful service!'),
(6, "It's okay, but definitely not worth the premium price tag. I expected more."),
(7, "Beautiful product, but much smaller than expected, especially for the price. The dimensions were listed, but without a point of reference (like a common object in the photo), it was hard to gauge the actual size"),
(8, "With extreme excitement, I opened up the packet. There are 6 shirts but I was so stunned and numb. The product I received was all XXL size even if I ordered a M size. I'm very disappointed with the incorrect item sent"); 

select * from datasciencebasics.basic.product_reviews;
--- ### AI functions on this table ### -- 

-- analye sentiment 
select review, ai_analyze_sentiment(review) as sentiment 
from datasciencebasics.basic.product_reviews; 

-- classify reviews
select review, 
    ai_classify(review, array(
      "Arrives too late",
      "Wrong size",
      "Wrong color",
      "Dislike the style"
    )) as reason  -- classify reviews into each group
from datasciencebasics.basic.product_reviews
where ai_analyze_sentiment(review) = 'negative';

-- extract info from reviews -- 
select 
  review, 
  ai_extract(review, array("usual size")) as usual_size, 
  ai_classify(review, array("Size is wrong", "Size is right"))
from datasciencebasics.basic.product_reviews;

-- generate responses with recommendations 
select 
  review, 
  ai_gen("Genereate a reply in 60 words to address the customer's review. Mention their opinions are valued and a 30% discount coupon code has been sent to their email. Customer's review: " || review)
from datasciencebasics.basic.product_reviews
where ai_analyze_sentiment(review) = 'negative';
```
